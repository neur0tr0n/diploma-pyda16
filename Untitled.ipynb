# Парсинг файла с логами FX-агрегатора

```python
import re
import matplotlib as plt
import numpy as np

directory = 'C:\\Users\\boris\\Downloads\\'
log_file = open(directory + 'fxaggregatorFX-quote.log', 'r')
fix_file = open(directory + 'quote.log', 'w')
quotes = {}
fix_list = []
quote_pear = 'eur/rub'
time_start = '20201127-08:15'
time_end = '20201127-08:20'

reg_exp = re.compile(r'(52=\d{8}-\d{2}:\d{2}:\d{2}\.\d{3}).*(269=\d).*(55=' + quote_pear + '_TOD).*(270=\d{1,3}\.\d{4}).*(271=\d{6,7})', re.IGNORECASE)
reg_exp_time = re.compile(r'\d{8}-\d{2}:\d{2}:\d{2}\.\d{3}')
for line in log_file:
    if reg_exp.search(line):
        out = reg_exp.search(line)
        current_time = reg_exp_time.search(out.group(1)).group(0)
        if time_start < current_time < time_end:
            print(current_time, out.groups())
```
