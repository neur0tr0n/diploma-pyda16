```python
def get_all_links(url_, search_string, page_count):   
    links = []
    params = {
        'search_query': search_string
    }
    for i in range(page_count):
        params['page'] = i
        res = requests.get(url_, params=params)
        time.sleep(0.2) 
        soup = bs(res.text, 'html.parser')
        news_blocks = soup.find_all('div', class_='search_results_item')
        article_introes = list(map(lambda x: x.find('div', class_='article_intro'), news_blocks))
        hrefs = list(map(lambda x: x.find('a').get('href'), article_introes))
        links += list(map(lambda x: 'https://www.kommersant.ru/' + x, hrefs))
    return links
    
url = 'https://www.kommersant.ru/search/results'
all_links = get_all_links(url, 'python', 5)
```

```python
import pandas as pd 

for link in all_links:
    article = requests.get(link)
    time.sleep(0.2)
    soup = bs(article.text, 'html.parser')
    if soup.find('h2', class_='article_name'): 
        title = soup.find('h2', class_='article_name').text
    else:
        title = soup.find('h1', class_='article_name').text
    if soup.find('div', class_='b-article__publish_date'):
        date = pd.to_datetime(soup.find('div', class_='b-article__publish_date').find('time').get('datetime'), dayfirst=True).date()
    elif soup.find('time', class_='title__cake'):
        date = pd.to_datetime(soup.find('time', class_='title__cake').get('datetime'), dayfirst=True).date()
    article_text = soup.find('div', class_='article_text_wrapper').text
    
    print(date, title)
    print(article_text)
```

```python
def get_news(links):
    news = pd.DataFrame()
    for link in links:
        article = requests.get(link)
        time.sleep(0.2)
        soup = bs(article.text, 'html.parser')
        if soup.find('h2', class_='article_name'): 
            title = soup.find('h2', class_='article_name').text
        else:
            title = soup.find('h1', class_='article_name').text
        if soup.find('div', class_='b-article__publish_date'):
            date = pd.to_datetime(soup.find('div', class_='b-article__publish_date').find('time').get('datetime'), dayfirst=True).date()
        elif soup.find('time', class_='title__cake'):
            date = pd.to_datetime(soup.find('time', class_='title__cake').get('datetime'), dayfirst=True).date()
        article_text = soup.find('div', class_='article_text_wrapper').text
        row = {'date': date, 'title': title, 'text': article_text}
        news = pd.concat([news, pd.DataFrame([row])])
    return news

all_news = get_news(all_links)
all_news
```
