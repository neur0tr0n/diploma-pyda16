```python
import requests
from bs4 import BeautifulSoup as bs
import re
import json

url = 'http://ftp.moex.com/pub/ClientsAPI/ASTS/Bridge_Interfaces/Currency/BrokerRisk33_vs_BrokerRisk32_Russian.html'
obj_name = re.compile(r'TABLE (\w+)')
params = {
    'Content-Type': 'text/html',
    'content': 'text/html; charset=utf-8'
}
```

```python
resp = requests.get(url, params=params)
resp_decoded = resp.content.decode('utf-8')
soup = bs(resp_decoded, 'html.parser')
obj_list = []
links = soup.find_all('a', href=True, text=True)
for link in links:
    if '#nowhere' in link['href']:
        obj_list.append(obj_name.findall(link.text))
obj_list = list(filter(None, obj_list))
print(obj_list)


#print(table_list)
# for table in table_list:
#     print(table_name.search(table)[0])
#table_description = soup.find_all('ul')
#print(table_description)

# for table in obj_list:
#     if table[0] == 'TABLE':
#         print(table[1]) 
```
